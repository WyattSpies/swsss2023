```python
from scipy.optimize import fsolve
import numpy as np
import matplotlib.pyplot as plt

#right hand side
def f(x):
    return np.exp(x) - 4*x

#initial guess
x0 = 1.4
#solve statement assigns solution to x
x = fsolve(f, x0)
```

```python
print(x)
```

```python
t = np.linspace(-3,3)
line = np.zeros(len(t))
```

```python
plt.plot(t, f(t))
plt.plot(t, line)
plt.show
```
